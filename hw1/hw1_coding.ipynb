```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy.stats import uniform, norm
from scipy.linalg import lstsq
```

```python
x = [0, 2, 3, 4]
y = [1,3,6,8]
m = 62/35
b = 18/35


xx = np.linspace(-1, 5, 1000)

plt.figure(figsize = (5,4))

plt.scatter(x, y, c = 'xkcd:salmon', label = 'Data')
plt.plot(xx, m*xx + b, lw = 3, label = 'Normal Equation Fit', c = 'xkcd:brick red')
plt.legend()
plt.show()
```

### Part (d)

```python
#parameters
x = [0, 2, 3, 4]
y = [1,3,6,8]
m = 62/35
b = 18/35

#set up figure
xx = np.linspace(-1, 5, 1000)
plt.figure(figsize = (5,4))

#make random data
r_x = uniform.rvs(-1, 6, size = 100)
r_y = m*r_x + b + norm.rvs(size = 100)

#plot random data
plt.scatter(r_x, r_y, c = 'xkcd:baby blue', label = 'Random Data')

#calculate and plot random fit
M = r_x[:, np.newaxis]**[0, 1]
p, res, rnk, s = lstsq(M, r_y)
plt.plot(xx, p[0] + xx *p[1], lw = 3, c = 'xkcd:blue', label = 'Random Data Fit')

#plot normal equation
plt.plot(xx, m*xx + b, lw = 3, label = 'Normal Equation Fit', c = 'xkcd:brick red')
#plot data
plt.scatter(x, y, c = 'xkcd:salmon', label = 'Data')

#the end
plt.legend()
plt.show()
```
